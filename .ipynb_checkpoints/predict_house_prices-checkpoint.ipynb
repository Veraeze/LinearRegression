```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#load the data

houses = pd.read_csv('datasets/HousingData.csv')
```

```python
# inspect for missing values

print(houses.isnull().sum())
```

```python
# drop rows with missing values

houses_clean = houses.dropna()
```

```python
# check data types

print(houses_clean.dtypes)
```

```python
# checking for outliers

print(houses_clean.describe())
```

```python
# data visualization

plt.figure(figsize=(12, 10))
sns.heatmap(houses_clean.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('correlation heatmap')
plt.show()
```

```python
# Separate features and target from the cleaned DataFrame for first model

X = houses_clean.drop('MEDV', axis=1)
y = houses_clean['MEDV']
```

```python
# split the dataset into the training set and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# fit multiple linear regression to the training set

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#predict the test set results

y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
# plot the actual vs predicted values

plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.7, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs. Predicted House Prices')
plt.grid(True)
plt.show()
```

```python
#calculate the coefficients

print(regressor.coef_)
```

```python
#calculate the intercept

print(regressor.intercept_)
```

```python
#calculate the R squared value

from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

```python
# train a new model because the R2 is 58% which means the model is not so accurate using selected features that have the most correlation based on the heatmap plotted earlier

selected_features = ['RM', 'LSTAT', 'PTRATIO', 'ZN']
X = houses_clean[selected_features]
y = houses_clean['MEDV']
```

```python
# split the dataset into the training set and test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
#  train a new model because the R2 is 54% which means the model is not so accurate.apply polynomial features and train selected features

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

selected_features = ['RM', 'LSTAT', 'PTRATIO']

poly_model = make_pipeline(PolynomialFeatures(degree=2, include_bias=False), LinearRegression())
poly_model.fit(X_train[selected_features], y_train)
y_pred = poly_model.predict(X_test[selected_features])
print(r2_score(y_test, y_pred))
```

```python
#  train a new model because the R2 is 60% which means the model is not so accurate.apply polynomial features and train all features

poly_model = make_pipeline(PolynomialFeatures(degree=2, include_bias=False), LinearRegression())
poly_model.fit(X_train, y_train)
y_pred = poly_model.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
# train a new model because the R2 is 60% which means the model is not so accurate. apply Polynomial Regression with Standardization

from sklearn.preprocessing import StandardScaler

poly_model = make_pipeline(StandardScaler(), PolynomialFeatures(degree=2, include_bias=False), LinearRegression())
poly_model.fit(X_train, y_train)
y_pred = poly_model.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
# reduce overfitting and improve generalization using ridge regression. train a new model.

from sklearn.linear_model import Ridge

poly_model = make_pipeline(StandardScaler(), PolynomialFeatures(degree=2, include_bias=False), Ridge(alpha=1.0))
poly_model.fit(X_train, y_train)
y_pred = poly_model.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
# reduce overfitting and improve generalization using lasso regression. train a new model.

from sklearn.linear_model import Lasso

poly_model = make_pipeline(StandardScaler(), PolynomialFeatures(degree=2, include_bias=False), Lasso(alpha=1.0))
poly_model.fit(X_train, y_train)
y_pred = poly_model.predict(X_test)
print(r2_score(y_test, y_pred))
```
