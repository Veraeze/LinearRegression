```python
# import necesaary packages

import numpy as np
import pandas as pd
```

```python
# load the data file 

loans = pd.read_csv('datasets/Loans_Dataset.csv')
```

```python
# analyze the dataset

print("dataset length: ", len(loans))
print("dataset shape: ", loans.shape)
print("dataset: \n", loans.head())
```

```python
# seperate the target variable

X = loans.drop(columns=["result", "sum"])
y = loans["result"]
```

```python
# split dataset into training and test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# train classifier using entropy

from sklearn.tree import DecisionTreeClassifier

entropy_clf = DecisionTreeClassifier(criterion="entropy", random_state=100,
                                max_depth=3, min_samples_leaf=5)
entropy_clf.fit(X_train, y_train)
```

```python
# make precitions using the test set 

y_pred = entropy_clf.predict(X_test)
y_pred
```

```python
# check accuracy using accuracy score

from sklearn.metrics import accuracy_score

print("accuracy is "), accuracy_score(y_test, y_pred)*100
```
